```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, Flatten, MaxPooling2D
from tensorflow.keras import Input
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing.image import load_img
from tensorflow.keras.preprocessing.image import img_to_array
from keras.preprocessing import image
import numpy as np
from PIL import Image
import cv2
import matplotlib.pyplot as plt
```

```python
Training_Path = '../input/pneumonia-dataset'
```

```python
train_datagen = ImageDataGenerator(rescale = 1./255, validation_split=0.20)
```

```python
train_data = train_datagen.flow_from_directory(Training_Path, 
                                               target_size=(150, 150),
                                               batch_size=32,
                                               color_mode="grayscale",
                                               class_mode='binary',
                                               subset = 'training')
val_data = train_datagen.flow_from_directory(Training_Path, 
                                           target_size=(150,150),
                                           batch_size=32, 
                                           color_mode="grayscale",
                                           class_mode='binary',
                                           subset = 'validation')
```

```python
train_data.class_indices
```

```python
model = Sequential()
model.add(Conv2D(128, kernel_size=3, activation='relu', input_shape=(150,150,1)))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(128, kernel_size=3, activation='relu'))
model.add(MaxPooling2D(2,2))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss = 'binary_crossentropy',
             optimizer = 'adam',
             metrics=['accuracy'])
```

```python
model_fit = model.fit(train_data,
                      validation_data  = val_data,
                      epochs=5)
```

```python
#im = Image.open('../input/pneumonia-dataset/Normal/Normal/0010f549-b242-4e94-87a8-57d79de215fc.jpg').convert('L')
im = Image.open('../input/pneumonia-dataset/Pnu/Pnu/020a16e3-baf9-4cf0-859c-c79b5253d717.jpg').convert('L')
im = im.resize((150, 150)) # Resizing to neccessary size. 
img = np.array(im).reshape(1, 150, 150) 
x = np.asarray(img)
x = 2*x/255 -1
x = np.transpose(x)
x = np.expand_dims(x,axis = 0)
images = np.vstack([x])
```

```python
#im = image.load_img('../input/pneumonia-dataset/Pnu/Pnu/019d950b-dd38-4cf3-a686-527a75728be6.jpg', target_size = (150,150,3))
plt.imshow(im)
plt.show()
classes = model.predict(images)
#print(classes)
max_index = np.argmax(classes[0])
val = classes[0][max_index]
if val <0.95:
    print('Normal with confidence', 1 - classes[0][max_index])
else:
    print('Pneumonia with confidence', classes[0][max_index])

#print('\n', dictionary[max_index])
```
